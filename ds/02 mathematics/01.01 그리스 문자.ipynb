```
$$ \alpha = \beta $$
```

라고 쓰면 다음처럼 독립된 문단으로 수식이 표시된다.

$$ \alpha = \beta $$

```{admonition} 연습 문제 1.1.2

주피터 노트북을 하나 만들고 인라인 수식과 디스플레이 수식으로 알파부터 오메가까지 써본다.
```
